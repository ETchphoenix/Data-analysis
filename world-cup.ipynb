Now we got all the teams that participate in the world cup

Now let's get rid of the possision issue to covert it to float

```python
for n in df.index:
    df['possession team1'].loc[n] = df['possession team1'].loc[n].replace('%', '')
    df['possession team2'].loc[n] = df['possession team2'].loc[n].replace('%', '')
    df['possession in contest'].loc[n] = df['possession in contest'].loc[n].replace('%', '')
df
```

```python
df['possession team1'] = df['possession team1'].astype('float')
df['possession team2'] = df['possession team2'].astype('float')
df['possession in contest'] = df['possession in contest'].astype('float')
df['possession team1'] = df['possession team1']/100
df['possession team2'] = df['possession team2']/100
df['possession in contest'] = df['possession in contest']/100
df.head()
```

Now get rid of unnecessary things

```python
df = df.drop(['date','hour','category'],axis=1)
df.head()
```

Get possession average for each nation team 

First replace nan with 0

```python
new_df = new_df.fillna(0)
new_df
```

```python
c = new_df.columns[1:] # columns to fill
for i in new_df.index:
    team1 = df[df['team1'] == new_df['team'].loc[i]]
    team2 = df[df['team2'] == new_df['team'].loc[i]]
    for co in c:
        if 'possession' in co:
            if 'team1' in co:
                new_df[co].loc[i] = team1[co].mean()
            else:
                new_df[co].loc[i] = team2[co].mean()
        else:
            if 'team1' in co:
                new_df[co].loc[i] = team1[co].sum()
            else:
                new_df[co].loc[i] = team2[co].sum()
new_df
```

```python
new_df['possession'] = (new_df['possession team1']+ new_df['possession team2'])/2
new_df.drop(['possession team1','possession team2'],axis=1, inplace=True)
c = new_df.columns[1:]
for co in c:
    if 'team1' in co: #to make it once
        new_co = co.replace(' team1','')
        new_df[new_co] = new_df[new_co+' team1']+ new_df[new_co+' team2']
        new_df.drop([new_co+' team1',new_co+' team2'],axis=1, inplace=True)
new_df
```

```python
#New columns
new_df[['win','loss','draw']] =[0,0,0]
```

```python
for i in new_df.index:
    team1 = df[df['team1'] == new_df['team'].loc[i]]
    team2 = df[df['team2'] == new_df['team'].loc[i]]
    for l in team1.index:
        if team1['number of goals team1'].loc[l] > team1['number of goals team2'].loc[l]:
            new_df['win'].loc[i] += 1
        elif team1['number of goals team2'].loc[l] > team1['number of goals team1'].loc[l]:
            new_df['loss'].loc[i] += 1
        else:
            new_df['draw'].loc[i] += 1
    for l in team2.index:
        if team2['number of goals team2'].loc[l] > team2['number of goals team1'].loc[l]:
            new_df['win'].loc[i] += 1
        elif team2['number of goals team1'].loc[l] > team2['number of goals team2'].loc[l]:
            new_df['loss'].loc[i] += 1
        else:
            new_df['draw'].loc[i] += 1
new_df['number_of_matches'] = new_df['win'] + new_df['loss'] + new_df['draw']
new_df
```
